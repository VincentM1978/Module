

## Fonction translate



>Crée une fonction translate qui traduit chacune des 5 valeurs possibles présente dans OPINION en français. Ta fonction doit prendre un str en paramètre, et retourner un autre str avec la traduction. Par exemple : translate("bad") doit retourner "mauvais". Idem pour very bad, not good not bad, good et very good. Tu peux utiliser un dictionnaire dict, ou bien des conditions imbriquées (le dictionnaire est plus élégant).

```
def translate(mot):
  """
  La fonction translate(mot) prend un mot relatant une OPINION en argument et
  renvoie la traduction française du mot.
  Si elle ne connaît pas le mot, elle renvoie le mot d'origine.
  """
  trad = {'very bad':'très mauvais', 'bad':'mauvais', 'not good not bad':'ni bon ni mauvais', 'good':'bon', 'very good':'très bon' }
  
  if mot in trad:
      return trad[mot]
  else:
      return mot

```

```
# application de la fonction translate à la série OPINION puis stockage dans la colonne OPINION
df2018['OPINION'] = df2018['OPINION'].apply(translate)
```

```
# création de la colonne MONTH contenant le numéro du mois
df2018['MONTH'] = df2018['DATE'].dt.month
```

```
# création de la colonne DAY contenant le numéro du jour
df2018['DAY'] = df2018['DATE'].dt.day
```

```
# conconténation des lignes du dataframe 2018 avec celles de 2019
df1819 = pd.concat([df2018, df_weather])
```

```
'''
Tu dois obtenir un DataFrame avec 730 lignes (365 * 2) et 25 colonnes :

20 colonnes communes
2 colonnes MONTH et DAY communes également
1 colonne flag dédiée à 2018
2 colonnes dédiées à 2019 : WEATHER_CODE_MORNING et WEATHER_CODE_NOON
'''
df1819.info()
```

```
# conversion de la colonne DATE en format datetime
df1819['DATE'] = pd.to_datetime(df1819['DATE'])
# création de la colonne YEAR avec application d'une fonction lambda
df1819['YEAR'] = df1819['DATE'].apply(lambda x: x.year)
```

```
import seaborn as sns
import matplotlib.pyplot as plt

# création de la figure avec deux lignes et deux colonnes
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

# histogramme de la vitesse du vent
sns.histplot(data = df1819, x = 'WINDSPEED_MAX_KMH', hue = 'YEAR', ax=axes[0,0])
axes[0,0].set_title('Histogramme de la vitesse du vent')

# boxplots pour les précipitations
sns.boxplot(data = df1819, x = 'YEAR', y = 'PRECIP_TOTAL_DAY_MM', ax=axes[0,1])
axes[0,1].set_title('Précipitations 2018 et 2019')

# scatterplots pour SUNHOUR et date
sns.scatterplot(data = df1819, x = 'DATE', y = 'SUNHOUR', hue = 'YEAR', ax=axes[1,0])
axes[1,0].set_title("Temps d'ensoleillement par jour, 2018 et 2019")

# violinplots pour la température maximale
sns.violinplot(data = df1819, x = 'YEAR', y = 'MAX_TEMPERATURE_C', ax=axes[1,1])
axes[1,1].set_title("Température maximale journalière, 2018 et 2019")

# ajustement de l'espacement entre les sous-graphiques
plt.tight_layout()

# affichage de la figure
plt.show()
```
